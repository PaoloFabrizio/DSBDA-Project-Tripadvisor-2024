```python
#!/usr/bin/env python
# coding: utf-8

import pandas as pd
from pymongo import MongoClient

client = MongoClient('localhost', 27017)

db = client['DB_Proj_DSBDA']
collection = db['ProjectDSBDA']

# mongo can't handle more than this number of rows
cursor = collection.find({}).limit(850000)

df = pd.DataFrame(list(cursor))

client.close()

# now dataset is loaded in a pandas dataframe, we can work in a sandbox enviroment
print(df.head(20))
print(df.shape)
```
