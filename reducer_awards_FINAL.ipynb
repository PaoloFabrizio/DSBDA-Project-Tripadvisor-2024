```python
"""reducer.py"""

import sys

current_id = None
current_restaurant = None
current_country = None
current_province = None
current_city_name = None
current_price_range = None
current_cuisines = None
current_features = None
current_veget = None
current_vegan = None
current_gluten = None
current_open_hours = None
current_avg_rating = None
count = 0

for line in sys.stdin:
    line = line.strip()
    splits = line.split('\t')

    # lines with != 14 fields are ignored
    if len(splits) != 14:
        continue
    
    id = splits[0]
    restaurant = splits[1]
    country = splits[2]
    province = splits[3]
    city_name = splits[4]
    award = int(splits[5])
    price_range = splits[6]
    cuisines = splits[7]
    features = splits[8]
    veget = splits[9]
    vegan = splits[10]
    gluten = splits[11]
    open_hours = splits[12]
    avg_rating = splits[13]
    
    if current_id != id:
        if current_id is not None:
            print(f"{current_id},{current_restaurant},{current_country},{current_province},{current_city_name},{count},{current_price_range},{current_cuisines},{current_features},{current_veget},{current_vegan},{current_gluten},{current_open_hours},{current_avg_rating}")

        current_id = id
        current_restaurant = restaurant
        current_country = country
        current_province = province
        current_city_name = city_name
        current_price_range = price_range
        current_cuisines = cuisines
        current_features = features
        current_veget = veget
        current_vegan = vegan
        current_gluten = gluten
        current_open_hours = open_hours
        current_avg_rating = avg_rating
        count = award
    else:
        count += award

if current_id is not None:
    print(f"{current_id},{current_restaurant},{current_country},{current_province},{current_city_name},{count},{current_price_range},{current_cuisines},{current_features},{current_veget},{current_vegan},{current_gluten},{current_open_hours},{current_avg_rating}")
```
